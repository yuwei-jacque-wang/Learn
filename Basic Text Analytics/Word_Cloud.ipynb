## 1. Prepare data for WordCloud

```python
# !pip install wordcloud

# Read file (course descriptions)
# Output of reading is a string (of entire text)
with open("data/Course-Descriptions.txt", 'r') as f:  
    filedata = f.read() 
    
# Create Stopword List (default from package):
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)

# Generate wordcloud 
wordcloud = WordCloud(stopwords=stopwords, max_words=25, \
                      background_color="white").generate(filedata)
```

## 2. Display WordCloud

```python
import matplotlib.pyplot as plt
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

## 3. Enhance WordCloud
- Add more stop words (to ignore)
- Limit number of words

```python
stopwords.update(['many','using','want','value'])

wordcloud = WordCloud(stopwords=stopwords, max_words=10, \
                      background_color="azure").generate(filedata)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
