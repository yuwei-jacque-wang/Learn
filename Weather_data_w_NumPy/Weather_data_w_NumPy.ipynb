# Select observations

We only want data lines with "GSN" and want to ignore the rest, so here we parse the lines

```python
stations = {} #assign a dictionary for station information

for line in open('stations.txt','r'): #each line can be entered into a loop as string in txt file
    if "GSN" in line:  
        fields = line.split() #split by whitespace by default, split() will parse a string into an array
        
        stations[fields[0]] = ' '.join(fields[4:]) #first item as the key of the station, 5th and after as name (join together)
```

Use numpy genfromtxt to load data into numpy array

```python
def parsefile(filename):
    return np.genfromtxt(filename,
                         delimiter = dly_delimiter,
                         usecols = dly_usecols,
                         dtype = dly_dtype,
                         names = dly_names)
```

```python
dly_delimiter = [11,4,2,4] + [5,1,1,1]*31 
  #from description of readme.txt, see the number of columns according with variable
    
dly_usecols = [1,2,3] + [4*i for i in range(1,32)] 
  #indicate columns we will use, so only (col 0 for ID is skipped) year, month, element and Value(n), each value represent a day
    
dly_dtype = [np.int32, np.int32, (np.str_,4)] + [np.int32] * 31

dly_names = ['year','month','obs'] + [str(day) for day in range(1, 31+1)] 
  #labels of columns used
```

We organize this into a better numpy format

```python
def unroll(record):
    startdate = np.datetime64('{}-{:02}'.format(record['year'],record['month']))
    dates = np.arange(startdate, startdate + np.timedelta64(1,'M'), np.timedelta64(1,'D'))
    
    rows = [(date,record[str(i+1)]/10) for i, date in enumerate(dates)] #enumerate() will assign an index for each observation
    return np.array(rows,dtype=[('date','M8[D]'),('value','d')])
```

Finally we try to concatenate all months in the filie into a single NumPy record array. We also let it allow us to choose observation (max or min)

```python
def getobs(filename, obs):
    return np.concatenate([unroll(row) for row in parsefile(filename) if row[2] == obs])

#The function will return the a time series of observations. 
```

# Clean missing data

We start by modifying getobs() function to assgin -999.9 (missing values) 

```python
def getobs(filename, obs):
    data = np.concatenate([unroll(row) for row in parsefile(filename) if row[2] == obs])
    data['value'][data['value'] == -999.9] = np.nan #replace -999.9 with "not a number"
    
    return data
```

We can observe missing data are cleaned by plotting the series

```python
lihue_tmax = getobs('USW00022536.dly', 'TMAX')
lihue_tmin = getobs('USW00022536.dly', 'TMIN')
pp.plot(lihue_tmax['date'],lihue_tmax['value'])
pp.plot(lihue_tmin['date'],lihue_tmin['value'])
```

Interpolation for nan values

```python
def fillnans(data):
    dates_float = data['date'].astype(np.float64) #np.astype() helps tranform data types
    
    nan = np.isnan(data['value'])
      #np.isnan() returns boolean masks for points that are indeed nan. 
      #So "nan" is a series containing T/F for each obs in data
        
    data['value'][nan] = np.interp(dates_float[nan], dates_float[~nan], data['value'][~nan]) 
      #"~" denotes negation for Boolean values 
```

```python
#Digression: an example to explain how interpolation works
x = np.array([1,3,5,7], 'd')
y = np.array([10,5,2,7], 'd')

xs = np.linspace(1,7) #x values for the interpolated values to land on 
ys = np.interp(xs, x, y) #1. the interpolation location, 2. original x value 3. original y value
pp.plot(xs, ys, '.')
pp.plot(x,y,'o')
```

# Smoothing data
  Convert to local average to smooth the noise. We firstly plot the orginal time series and then smoothed series.

```python
pp.plot(lihue_tmin['date'], lihue_tmin['value'])
```

```python
def plot_smoothed(data, win=10): #win is window length for computing average
    smoothed = np.correlate(data['value'], np.ones(win)/win, 'same')
      #np.correlate is to compute cross-correlation for two series. 
      #Correlating with [0.1, 0.1 ...0.1] generates mean of win length.
    
    pp.plot(data['date'], smoothed)
    
plot_smoothed(lihue_tmin)
```

# Computing daily records

Define a function that allows us to select data from a certain year

```python
def selectyear(data, year):
    start = np.datetime64('{}'.format(year))
    end = start + np.timedelta64(1, 'Y')
    
    return data[(data['date'] >= start) & (data['date'] < end)]['value']
        #This is a very useful boolean masks with logical operator for selecting range of data
    
selectyear(lihue_tmin, 2002)
```

Stack data of each year into one single vector, by stacking data from each selected year

```python
lihue_tmin_all = np.vstack([selectyear(lihue_tmin, year)[:365] for year in range(1951, 2014+1)])
  # [:365] must be added, otherwise data are not of the same dimension because of leep years
    
lihue_tmax_all = np.vstack([selectyear(lihue_tmax, year)[:365] for year in range(1951, 2014+1)])
    
print(lihue_tmin_all)
print(lihue_tmin_all.shape)
```
